```python
import geopandas as gpd
import pandas as pd
from datetimerange import DateTimeRange
import branca.colormap as cm
import xyzservices.providers as xyz
```

```python
df=pd.read_csv('data6.csv')
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.lon, df.lat,crs="EPSG:4326")).to_crs('epsg:2100')
```

```python
step = cm.StepColormap(['red','blue','green'],
                       vmin=0, vmax=30, index=[0,5,10,30],
                       caption='Speed')

gdf.explore(column='vessel_id',cmap='Dark2',tiles=xyz.CartoDB.Voyager,popup=True)
```

```python
gdf.dt=gdf.dt.apply(lambda x: pd.to_datetime(x))
gdf.dt_min=gdf.dt_min.apply(lambda x: pd.to_datetime(x))
gdf.dt_max=gdf.dt_max.apply(lambda x: pd.to_datetime(x))
```

```python
pairs={
(gdf.vessel_id[i],gdf.vessel_id[j])
for i in range(len(gdf)) 
for j in range(len(gdf)) 
if (gdf.geometry[j].distance(gdf.geometry[i])/1852)<=0.5 
if gdf.vessel_id[i]!=gdf.vessel_id[j]
if DateTimeRange(gdf.dt_min[i],gdf.dt_max[i]).is_intersection(DateTimeRange(gdf.dt_min[j],gdf.dt_max[j]))
}
```

```python
print(f'Number of unique pairs of vessel_ids: {len(pairs)}')
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
for i,id in enumerate(pairs):
    print(i+1,id)
    if i+1==5:
        break
```

```python
from itertools import chain
ships=set(list(chain(*pairs)))

print(f'Number of unique vessel_ids: {len(ships)}')
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
for i,id in enumerate(ships):
    print(i+1,id)
```
